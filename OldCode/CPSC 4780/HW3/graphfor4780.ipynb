```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "BLOCK_SIZE": [1, 2, 4, 8, 10, 20],
    "Execution_Time": [.012288, .013536, .012032, .013216, .011424, .013312]  
}

# Convert to DataFrame
df = pd.DataFrame(data)
```

```python
# Plotting the data
plt.figure(figsize=(8, 5))
plt.plot(df["BLOCK_SIZE"], df["Execution_Time"], marker='o', linestyle='-', label="Execution Time")
plt.title("Execution Time vs. BLOCK_SIZE")
plt.xlabel("BLOCK_SIZE")
plt.ylabel("Execution Time (ms)")
plt.grid()
plt.legend()
plt.show()
```
